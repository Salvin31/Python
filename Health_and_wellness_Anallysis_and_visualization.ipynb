```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Python group 6 dataset.xlsx")
```

```python
# Display basic info
print("Dataset Info:")
df.info()
```

```python
# Display first few rows
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Display last few rows
print("\nLast 5 Rows:")
print(df.tail())
```

```python
# Display entire dataset overview
print("\nFull Dataset:")
print(df.to_string())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Check duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# Data types of each column
print("\nData Types:")
print(df.dtypes)
```

```python
# Count unique values per column
print("\nUnique Values per Column:")
print(df.nunique())
```

```python
# Value counts for categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"\nValue Counts for {col}:")
    print(df[col].value_counts())
```

```python
# Detect outliers using IQR
numerical_columns = df.select_dtypes(include=['number']).columns
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()
    print(f"\nOutliers in {col}: {outliers}")
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Values:")
print(df.duplicated().sum())
```

```python
plt.figure(figsize=(8, 6))
df.groupby("Product line")["Total_Cost"].sum().plot(kind="pie", autopct='%1.1f%%', cmap='viridis', startangle=90)
plt.title("Sales Distribution by Product Line")
plt.ylabel('')  # leaving y-label empty for better display
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.groupby("Date")["Total_Cost"].sum().plot(kind="line", marker='o', color='b')
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["Total_Cost"], bins=30, kde=True, color='green')
plt.title("Distribution of Total Cost")
plt.xlabel("Total Cost ($)")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
newdata = df[["Payment","Total_Cost"]]
newdata
```

```python
bar_data = newdata.groupby("Payment").sum().reset_index()
bar_data.columns = ["Payment Mode","Sum of Total Cost"]
bar_data
```

```python
#Bar Chart
plt.figure(figsize=(3,2))
plt.bar(bar_data["Payment Mode"],bar_data["Sum of Total Cost"],color = ["Blue","Red","Green"])
plt.title("The Bar Chart for Payment wise Summation of total Cost")
plt.xlabel("Payment Mode")
plt.ylabel("Total Cost")
plt.show()
```

```python
newdata2 = df[["Product line","Quantity"]]
newdata2
newdata_2= newdata2.groupby("Product line").sum().reset_index()
newdata_2.columns = ["Product Line Name","Total no. of Quantity"]
newdata_2
```

```python
#Scatter Plot
plt.figure(figsize=(14,4))
plt.scatter(newdata_2["Product Line Name"],newdata_2["Total no. of Quantity"])
plt.title("The Scatter Plot for each Product Line and Total No. of quantity sold")
plt.xlabel("The Product Line")
plt.ylabel("Total Quantity")
plt.show()
```
